```python
# splitting data into train test

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,stratify=y)
```

```python
x_train.head(2)  # we need to convert our data into numeric form to perform any model
```

```python
x_test.head(2)
```

```python
# text encoding

from sklearn.feature_extraction.text import CountVectorizer
vectroizer = CountVectorizer(stop_words="english")
x_train = vectroizer.fit_transform(x_train["status"])
x_test = vectroizer.transform(x_test["status"])
```

```python
x_train
```

```python
# above is the sparse array to see this saprse array we need to convert into dataframe
pd.DataFrame(x_train.toarray())
                                # here you may have observed that there is no names for any column so we need to change the name
                                # for which name what is assigned
```

```python
x_train = pd.DataFrame(x_train.toarray(),columns=vectroizer.get_feature_names())
x_train
```

```python
x_test
```

```python
x_test = pd.DataFrame(x_test.toarray(),columns=vectroizer.get_feature_names())
x_test
```

```python
# training our training data
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
